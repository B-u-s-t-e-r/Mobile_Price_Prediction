# Model Building

```python
X = df.drop('Price',axis=1)
y = df['Price']
```

```python
print(X)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size = 0.2, random_state = 7)
```

# scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
```

```python
lr  = LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
print(f"R2_Score is: {r2_score(y_test,y_pred)}")
print(f"Mean Absolute error is: {mean_absolute_error(y_test,y_pred)}")
print(f"Mean Squared error is: {mean_squared_error(y_test,y_pred)}")
print(f"Root Mean Squared Error is: {np.sqrt(mean_squared_error(y_test,y_pred))}")
```

# XGB Regressor

```python
from xgboost import XGBRegressor
xgb = XGBRegressor()
xgb.fit(X_train,y_train)
```

```python
y_pred_xgb = xgb.predict(X_test)
y_pred_xgb
```

```python
print(f"R2_Score is: {r2_score(y_test,y_pred_xgb)}")
print(f"Mean Absolute error is: {mean_absolute_error(y_test,y_pred_xgb)}")
print(f"Mean Squared error is: {mean_squared_error(y_test,y_pred_xgb)}")
print(f"Root Mean Squared Error is: {np.sqrt(mean_squared_error(y_test,y_pred_xgb))}")
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeRegressor
dtree = DecisionTreeRegressor()
dtree.fit(X_train,y_train)
```

```python
y_pred_dtree = dtree.predict(X_test)
y_pred_dtree
```

```python
print(f"R2_Score is: {r2_score(y_test,y_pred_dtree)}")
print(f"Mean Absolute error is: {mean_absolute_error(y_test,y_pred_dtree)}")
print(f"Mean Squared error is: {mean_squared_error(y_test,y_pred_dtree)}")
print(f"Root Mean Squared Error is: {np.sqrt(mean_squared_error(y_test,y_pred_dtree))}")
```
